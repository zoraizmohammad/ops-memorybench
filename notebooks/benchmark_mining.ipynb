# Benchmark mining

This notebook shows how candidate benchmark tasks are mined from captured
trajectories. A good memory task is one where the answer depends on durable context
not inferable from the current app state, and where the effect of memory can be
judged with explicit evidence. The miner surfaces repeated corrections, recurring
workflows, and explicit durable statements, each with the trajectory evidence and a
rationale, for a human curator to promote into a task spec.

```python
from pathlib import Path

from ombench.eval.miner import mine_candidates
from ombench.traces.converters import from_claude_code_session

FIXTURES = Path('..') / 'fixtures' / 'trajectories'
runs = [from_claude_code_session(FIXTURES / 'claude_code_reschedule.jsonl')]
candidates = mine_candidates(runs, min_repeats=1)
for c in candidates:
    print(c.kind, '-', c.summary)
    print('   rationale:', c.rationale.strip()[:100])
```

## Curated benchmark

The fifteen curated tasks live in `benchmarks/tasks`. Each documents in its
`why_memory` field exactly why the snapshot alone is insufficient, which is the
filter that separates a good memory task from one solvable without memory.

```python
from ombench.eval.tasks import load_tasks

tasks = load_tasks(Path('..') / 'benchmarks' / 'tasks')
for t in tasks:
    print(f"{t.task_id}: {t.title}")
    print('   why memory:', t.why_memory.strip()[:90])
```
